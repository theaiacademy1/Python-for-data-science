## <span style="color: green"> Making a copy of your data </span>
* <span style="color: green">There might be situations where we need to work with a copy of our data without modifying the original dataset.</span>
*<span style="color: green"> We can create copies of the pandas DataFrame.</span>
* <span style="color: green">Python provides us with two ways to create copies:</span>
    * <span style="color: green">First is **shallow copy**</span>
    * <span style="color: green">Second one is **deep copy.**</span>
* <span style="color: green"> **Shallow copy** creates a new variable that shares the reference of the original object.It doesn't create a completely new object in memory </span>
* <span style="color: green"> To create a shallow copy</span>

    * <span style="color: green">Shallow_copy = df.copy(deep=False)</span>

    * <span style="color: green">Simply do: shallow_copy = df</span>

* <span style="color: green">Shallow copy creates a new variable that points to the same memory location as the original object. 
It doesn't create a brand new object in memory. So, any changes to the shallow copy will also be reflected in the original object, and vice versa.</span>
* <span style="color: green">Useful when you're working with mutable objects and want the changes to be reflected in both the original and the copy</span>

* <span style="color: red">**Deep Copy** creates a completely independent copy of the DataFrame, where changes made to the copy won't affect the original data.</span>
* <span style="color: red">To create a deep copy, you use the same `copy()` method, but you set the `deep` parameter to `True`:</span>
* <span style="color: red"> To create a Deep copy</span>
    *  <span style="color: red">deep_copy = cars_data.copy(deep=True)</span>
    * <span style="color: red">By default, `deep=True` when you use `copy()` without any arguments, so the above line is equivalent to: deep_copy = cars_data.copy()</span>
* <span style="color: red">Deep copy, a completely new object is created in memory, with no reference to the original object. Any changes you make to the deep copy will not affect the original DataFrame, and vice versa.</span>


```python
shallow_copy = df.copy(deep=False) # making a shallow copy of original Data 
shallow_copy.info()
```

## Data types. 

* <span style='color:green'>The way information is stored in a DataFrame or any Python object significantly impacts the analysis you perform on your data and the output of your calculations.
* <span style='color:green'>For example, you cannot perform numerical operations on a string, and similarly, you cannot perform string-related operations on numerical data. 
* <span style='color:green'>There are two main types of data: </span>
    * `Numeric`
    * `Character types.`
* <span style='color:green'>The Numeric data types include all numerical values, represented as integers and floats. </span>
* <span style='color:green'>Strings are known as objects that can store values containing both numbers and characters. </span>
* <span style='color:green'>Integer: In Pandas, it's called int64, and it's like a big 64-bit integer for handling large numbers.</span>
* <span style='color:green'>Floating-Point Number also called as float: Pandas calls it float64, and it's a 64-bit number that can handle decimals with precision.</span>
* <span style='color:green'>In Python, there are two primary data types that can handle character types: 'category' and 'object'.</span>
* <span style='color:green'>Categorical: This is handy for categorical data, like labels or categories, making storage and manipulation efficient.</span>
* <span style='color:green'>'Object' data types are useful when dealing with columns that have diverse data types or when you're not sure what type of data to expect.</span>
* <span style='color:green'>If you have a limited and fixed number of string values, using the 'category' data type is more memory-efficient. </span>
* <span style='color:green'>If you have a column with mixed types or many different string values, using the 'object' data type is more appropriate.</span>


### How data types are named in Python and how they are named in the Pandas library:
* <span style='color:red'>In base Python, integers are represented as 'int', while in the Pandas library, they are represented as 'int64'. The 'int64' data type corresponds to all numeric characters without decimal points.</span>
* <span style='color:red'> Next, we have floating-point numbers, represented as 'float' in Python and 'float64' in Pandas. The 'float64' data type corresponds to all numeric characters with decimal points.</span>
* <span style='color:red'>The number 64 in both refers to the memory allocated to store the data in each cell. This allocation determines how many digits can be stored in each cell. The maximum it can store is up to 64 bits, which is equivalent to 8 bytes.</span>


* In Python, there are two primary data types that can handle character types: 
    * category
    * object
* Let us discuss the difference between these two data types.</span>
* Categorical: This is handy for categorical data, like labels or categories, making storage and manipulation efficient.
* 'Object' data types are useful when dealing with columns that have diverse data types or when you are not sure what type of data to expect.
* A category data type is used for string values that consist of only a few different values. Converting such string variables to a categorical variable can save memory, as it avoids storing multiple identical string values. 
* Instead, categorical variables take on a limited and fixed number of possible values. Since there is a limit to the length, a categorical variable can only accommodate a fixed number of possible values. If you have many different values that form categories, it is better to use the 'object' data type.
* The object data type is assigned to a column when it has mixed data types, such as numbers represented as strings enclosed within single or double quotes or a mix of numbers and strings. For example, a column containing "0", "1", "category 1", "category 2", and "category 3" would be considered an 'object' data type. 
* In summary, if you have a limited and fixed number of string values, using the 'category' data type is more memory-efficient. 
* However, if you have a column with mixed types or many different string values, using the 'object' data type is more appropriate.

### Preparing Data for Binning: Adding a New Column

* To categorize car prices into bins, we need a place to store this information. We do this by adding a new column to our DataFrame intended for this categorical data:


```python
# Add a new column for storing price categories

df1.insert(loc=len(df1.columns), column='price_class', value='')
```

```python
df1.head()
```

### Implementing Binning with For Loops and Conditional Statements

* Use a combination of for loops and if-else statements to categorize each car's price

```python
for i in range(len(df1)):
    price = df1.loc[i, 'Price']  # Access the price for each car
    if price <= 8450:
        df1.loc[i, 'price_class'] = 'Low'  # Price is in the low range
    elif price <= 11950:
        df1.loc[i, 'price_class'] = 'Medium'  # Price is in the medium range
    else:
        df1.loc[i, 'price_class'] = 'High'  # Price is in the high range
```

### Defining the Conversion Function
* Function named `convert_to_years` that takes the age in months as input and returns the age in years. 
* To make the data cleaner, we’ll also round the result to one decimal place to avoid overly precise fractions of years that are hard to interpret.

```python
# Converts age from months to years, rounded to one decimal place.
def convert_to_years(months):
   
    years = months / 12
    return round(years, 1)
```

```python
### Applying the Function to the Dataset
df1['age_converted'] = df1['Age'].apply(convert_to_years)
```

```python
df1.head()
```

### Extending the Functionality
* Convert car ages and calculate how many kilometers a car has been driven per month.

```python
# Converts age and computes kilometers per month.
def detailed_conversion(months, total_kilometers):
    """Converts age and computes kilometers per month."""
    years = round(months / 12, 1)
    km_per_month = round(total_kilometers / months, 1) if months != 0 else 0
    return years, km_per_month
```

```python
df1['age_converted'], df1['km_per_month'] = zip(*df1.apply(lambda row: detailed_conversion(row['Age'], row['KM']), axis=1))
```

### Frequency table
* Pandas.crosstab()
* This function is used to compute a simple cross-tabulation of two (or more) factors. It is very useful for summarizing data to observe relationships between categorical variables.
* **index=df1['FuelType']**: This parameter specifies the index of the resulting table. Here, it uses the `FuelType` column from the DataFrame `df1`. Each unique value in this column will become a row in the resulting table.
* **columns='count**: Counts the occurrences of each unique value in the specified `index`.
* **dropna=True**: This parameter tells pandas to exclude any rows from the DataFrame that have missing values (`NaN`) in the `FuelType` column. This ensures that the frequency count only includes valid data.

```python
pd.crosstab(index=df1['FuelType'],columns='count', dropna=True)
```

### Two way table
* We will explore the frequency of distribution of **gearbox** types with respect to **fuel** types of the cars
* This way of analysing data helps us to not only understand the frequency of categories but also how these categories relate to others in our dataset. 

```python
df1.head()
```

* **index=df1['Automatic']** parameter sets the rows of the resulting table based on the `Automatic` column in the DataFrame `df1`. This column represents whether a car has an automatic gearbox which is 1 or a manual gearbox which is 0.
* **columns=df1['FuelType']** sets the columns of the table based on the `FuelType` column. Each unique value in this column will form a separate column in the resulting table.


```python
pd.crosstab(index=df1['Automatic'], columns=df1['FuelType'], dropna=True)
```

### Joint probability
* It's the likelihood of two independent events happening simultaneously. 
* For example, the probability of a car having a manual gearbox and using CNG as fuel.
* To calculate these joint probabilities using our two-way table, 
* add an extra parameter: `normalize=True`. 
* This parameter changes all the numerical counts in the table into proportions of the total, giving us the joint probabilities across our categories.


```python
pd.crosstab(index=df1['Automatic'], columns=df1['FuelType'], dropna=True, normalize=True)
```

### Marginal probabilities. 
* Marginal probability refers to the probability of a single event occurring, without considering the influence of other variables. 
* This is particularly useful when you want to understand the standalone impact or occurrence rate of one variable across different contexts.
* `normalize=True` converts the table values into proportions, and `margins=True` adds row sums and column sums to the table. 
* These sums are the marginal probabilities, showing the total probability distribution across either rows or columns, irrespective of the other variable.

```python
pd.crosstab(index=df1['Automatic'], columns=df1['FuelType'], dropna=True, normalize=True, margins=True)
```

### Conditional probabilities. 
* Conditional probability measures the likelihood of an event occurring, given that another specific event has already occurred. It helps us explore dependencies between variables more deeply.
* To calculate conditional probabilities with pandas, we still use the `pd.crosstab` function, but we'll adjust the normalization parameter to focus on one variable at a time. 
* This setup allows us to understand the probability of one event based on the occurrence of another.
* `normalize='index'`: This key parameter normalizes the row data to sum to 1.0, meaning it calculates the probability distribution of fuel types for each category of gearbox.


```python
pd.crosstab(index=df1['Automatic'], columns=df1['FuelType'], normalize='index', dropna=True)
```

### Normalizing through Columns
* How gearbox types vary with different fuel types we adjust our conditional probability calculation by normalizing across the columns of our two-way table. 
* This adjustment helps us understand the probability distribution of gearbox types given a specific fuel type.

```python
pd.crosstab(index=df1['Automatic'], columns=df1['FuelType'], normalize='columns', dropna=True)
```

### Correlation. 
* Correlation is a statistical measure that describes the strength and direction of a relationship between two numerical variables.
* **Positive Correlation:** As one variable increases, so does the other. For example, an increase in a car’s weight typically corresponds to an increase in its price.
* **Negative Correlation:** As one variable increases, the other decreases. For example, as the age of a car increases, its price tends to decrease.
* **No Correlation:** There is no discernible pattern in the changes between two variables.


#### Correlation values range from -1 to +1:
* Values close to +1 indicate a strong positive correlation.
* Values close to -1 indicate a strong negative correlation.
* A value of 0 indicates no correlation.
#### Visualizing Correlation with Scatter Plots:
* Scatter plots are useful for visually identifying the type of correlation between two variables. 
* A plot showing a clear upward trend indicates positive correlation, while a scattered plot with no clear pattern suggests little to no correlation.
#### Calculating Correlation in Python:
* To calculate correlations in Python, we use the `.corr()` method from pandas, typically on a DataFrame that only includes numerical data. 



```python
# Assuming our DataFrame containing both numerical and categorical data
numerical_data = df1.select_dtypes(exclude=['object'])

# Compute the correlation matrix
corr_matrix = numerical_data.corr(method='pearson')  # Default method is Pearson
print(corr_matrix)
```

### What are Bar Plots
* A bar plot is a graphical representation that uses bars to show comparisons among categories. It displays categorical data with rectangular bars, where each bar’s height is proportional to the value it represents. 
* Bar plots are particularly useful for representing frequency distributions of categorical variables, unlike histograms which are used for numerical variables.
****
#### Key Differences Between Bar Plots and Histograms

* **Bar Plots**: Separate the bars to represent different categories, which helps in highlighting the discrete nature of each category.
* **Histograms**: Typically connect the bars to show continuous data distributions.
****
#### When to Use Bar Plots

* Bar plots are ideal when you need to compare the frequency, count, or other measure (like mean) across different categories. They make it easy to see relative sizes of features among different categories.

- **Maximum (Upper Whisker)**: The upper whisker extends to 1.5 * IQR above the third quartile and represents the maximum non-outlier value.
- **Outliers**: These are represented as individual points beyond the whiskers. They indicate extreme values that may warrant further investigation.

### Understanding how to identify missing values
* It is an crucial first step in handling them effectively. 
* In pandas, missing values are usually represented by `NaN` (Not a Number). 
* This special value is used to denote missing or undefined values, typically arising from empty cells in your data files, such as CSVs. 
* When pandas encounter these empty cells during data import, it automatically converts them to `NaN`.
* To systematically check for these missing values in your pandas DataFrame, there are a couple of handy functions we can use: `isnull()` and `isna()`. 
* Both functions are essentially the same and can be used interchangeably to detect missing values across your DataFrame.
* These commands will return a new DataFrame where each cell is either `True` or `False`. 
* Cells with `True` indicate missing values (`NaN`), and those with `False` indicate non-missing values.

### Analyzing and Imputing Categorical Data
* when it comes to imputing missing values in categorical variables like 'Fuel type' and 'Met color', using the mode (the most frequently occurring category) is a practical and common approach. 
* To find out which category is the most frequent, you can use the `value_counts()` method. This will help you decide which category to use for filling in the missing values.

```python
# Get the frequency of each category in 'Fuel type'
df3['FuelType'].value_counts()
```

### Considerations for handling categorical data: 
* **Mode Imputation:** This is the simplest and often most effective method for categorical data, especially when there is a clear majority category. In this case, 'Petrol' is the mode as it appears most frequently. Imputing missing 'FuelType' values with 'Petrol' would be statistically justified given its high frequency, as it minimizes disruption to the existing distribution of data.
* **Consistency and Bias:** Imputing with the mode (i.e., 'Petrol') maintains the proportion and distribution of categories, reducing the potential bias that could be introduced if a less frequent category were used.
* **Practical Implications:** Since 'Petrol' is the most common fuel type, replacing missing values with this category also aligns with likely real-world scenarios, where a randomly selected car is more probable to be petrol-fueled given current data.


```python
# Impute missing values in 'FuelType' with the most frequent category
most_common_fuel_type = df3['FuelType'].mode()[0]  # This will be 'Petrol'
df3['FuelType'].fillna(most_common_fuel_type, inplace=True)
```
